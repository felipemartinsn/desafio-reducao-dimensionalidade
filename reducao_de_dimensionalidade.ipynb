```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
caminho_imagem = '10.jpg'
imagem = Image.open(caminho_imagem)
```

```python
plt.imshow(imagem)
```

```python
def grayscale_conversor(img):
  img_array = np.array(img)
  r, g, b = img_array[:, :, 0], img_array[:, :, 1], img_array[:, :, 2]
  gray = 0.299*r + 0.587*g + 0.114*b
  return gray
```

```python
imagem_cinza = grayscale_conversor(imagem)
```

```python
def binarizar_imagem(imagem, limiar):
    imagem_binaria = np.where(imagem_cinza < limiar, 0, 255).astype(np.uint8)
    return imagem_binaria
```

```python
limiar = 150 #valor arbitrario
imagem_binaria = binarizar_imagem(imagem_cinza, limiar)
```

```python
plt.subplot(1, 3, 1), plt.imshow(imagem, cmap='gray'), plt.title('Imagem Original')
plt.subplot(1, 3, 2), plt.imshow(imagem_cinza, cmap='gray'), plt.title('Escala CInza')
plt.subplot(1, 3, 3), plt.imshow(imagem_binaria, cmap='gray'), plt.title('Imagem Binarizada')
plt.subplots_adjust(wspace=0.5)
plt.show()
```
